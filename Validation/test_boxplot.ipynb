### Boxplot

```python
import os
import csv
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from statistics import variance



%matplotlib inline
```

```python
project = 'JUnit'
metrics = ['wmc', 'dit', 'noc', 'cbo', 'rfc', 'lcom*', 'tcc', 'fanin', 'fanout', 'totalMethodsQty', 'publicMethodsQty', 'totalFieldsQty', 'publicFieldsQty']
df = pd.read_csv(f'./MetricBugs/{project}.csv', sep=";", decimal=',')
i = 0
```

```python
project = os.listdir('./MetricBugs/')[8].replace('.csv','')
project
```

```python
dict_percentil_discrete = {
    'wmc' : {70: 7, 90: 22},
    'dit': {70: 2, 90: 3},
    'noc': {70: 4, 90: 9},
    'cbo': {70: 10, 90: 22},
    'rfc': {70: 9, 90: 29},
    'fanin': {70: 2, 90: 7},
    'fanout': {70: 7, 90: 15},
    'totalMethodsQty': {70: 5, 90: 12},
    'publicMethodsQty': {70: 3, 90: 9},
    'totalFieldsQty': {70: 2, 90: 5},
    'publicFieldsQty': {70: 2, 90: 7}
}

dict_percentil_continuo = {
    'lcom*': {70: 0.4, 90: 0.7917},
    'tcc': {70: 0.4483, 90: 1}
}
```

```python

df_copy = df.copy()
for key in dict_percentil_discrete.keys():
    arr = []
    for value in df_copy[key]:
        if value >= 0 and value <= dict_percentil_discrete[key][70]:
            arr.append('1 - Bom')
        elif value > dict_percentil_discrete[key][70] and value <= dict_percentil_discrete[key][90]:
            arr.append('2 - Regular')
        else:
            arr.append("3 - Ruim")
    
    df_copy[f'Intervalos de {key}'] = arr
    

for key in dict_percentil_continuo.keys():
    arr = []
    for value in df_copy[key]:
        if value >= 0 and value < dict_percentil_continuo[key][70]:
            arr.append('1 - Bom')
        elif value >= dict_percentil_continuo[key][70] and value < dict_percentil_continuo[key][90]:
            arr.append('2 - Regular')
        else:
            arr.append("3 - Ruim")
    
    df_copy[f'Intervalos de {key}'] = arr

df_copy
```

```python
for metric in metrics:    
    df_copy.boxplot(column=['bug'], by=f'Intervalos de {metric}', )
    plt.title('Boxplot de Bugs agrupados pelos Intervalos de ' + metric)
    plt.suptitle('')
    metric = metric.replace('*','')
    if not os.path.exists(f'./Boxplots/{metric}/'): 
        os.makedirs(f'./Boxplots/{metric}/')
    plt.savefig(f'./Boxplots/{metric}/{project}_boxplot_{metric}.png')
    plt.close()
```

### R2

```python
groups = ['1 - Bom', '2 - Regular', '3 - Ruim']

arr = []
for metric in metrics:
    n_total = 0
    aux = 0
    totalBugs = pd.Series([])
    for group in groups:
        bugs = df_copy[df_copy['Intervalos de '+metric] == group]
        bugs = bugs[bugs['bug'] >= 0]['bug']
        n_total += len(bugs)
        if(len(bugs)): aux += len(bugs)*variance(bugs)
        
        totalBugs = totalBugs.append(bugs)
        
    variance_mean = aux/n_total
    variance_total = variance(totalBugs)

    r2 = 1 - variance_mean/variance_total
    arr.append([metric, variance_mean, variance_total, r2])

df_r2 = pd.DataFrame(arr, columns=['Metrica', 'Variancia media', 'Variancia total', 'R2'])
df_r2.to_csv('./R2/Correlacao_'+project+'.csv', sep=';', decimal=',')
```
